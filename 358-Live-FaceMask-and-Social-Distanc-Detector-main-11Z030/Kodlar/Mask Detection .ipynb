```python
import numpy as np
#loading images of with_mask and without mask
without_mask = np.load('without_mask.npy')
with_mask = np.load('with_mask.npy')
with_mask =  with_mask.reshape(200,50 * 50* 3)
without_mask = without_mask.reshape(200,50 * 50*3)
print(with_mask.shape)
print(without_mask.shape)
all_img = np.r_[with_mask, without_mask] #concatinating data of wit_mask and without_mask
print(all_img.shape)
label = np.zeros(all_img.shape[0]) #taking zeros from images since. Here in (400,7500) 400 its images and 7500 are its dimensons
label[200:] = 1.0 #we are considering images of with mask as 1 (that are rest 200 images)
names = {0: 'Mask', 1: 'No mask'}
```

```python
from sklearn.svm import SVC #Support Vector Classification
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(all_img, label, test_size = 0.25)
print(xtrain.shape)
#applying ML
svc = SVC()
svc.fit(xtrain, ytrain)
#getting Y prediction from xtest
y_pred = svc.predict(xtest)
```

```python
import cv2
from scipy.spatial import distance as dist
#capturing the video from the user and making a rectangle on it
Haarcascade_data = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
video =  cv2.VideoCapture(0)
face_data = []
font = cv2.FONT_HERSHEY_COMPLEX
while True:
    flag, img = video.read()
    if flag == True:
        detect_face = Haarcascade_data.detectMultiScale(img)
        #Social Distancing Estimator
        no_of_faces = len(detect_face)
        face = cv2.putText(img, str(len(detect_face))+" Face", (60, 20), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0) , 2, cv2.LINE_AA) 
        stack_x = []
        stack_y = []
        stack_x_print = []
        stack_y_print = []
        global D
        
        if len(detect_face) == 0:
            pass
        else:
            for i in range(0, len(detect_face)):
                x1 = detect_face[i][0]
                y1 = detect_face[i][1]
                x2 = detect_face[i][0] + detect_face[i][2]
                y2 = detect_face[i][1] + detect_face[i][3]
                
                mid_x = int((x1 + x2)/2)
                mid_y = int((y1 + y2)/2)
                stack_x.append(mid_x)
                stack_y.append(mid_y)
                stack_x_print.append(mid_x)
                stack_y_print.append(mid_y)
                
                face = cv2.circle(img, (mid_x, mid_y), 3 , [255,0,0] , -1) 
                
            if len(detect_face) == 2:
                D = int(dist.euclidean((stack_x.pop(), stack_y.pop()), (stack_x.pop(), stack_y.pop())))
                face = cv2.line(face, (stack_x_print.pop(), stack_y_print.pop()), (stack_x_print.pop(), stack_y_print.pop()), [0,0,255], 2)
            else:
                D = 0
            
            if D<250 and D!=0:
                face = cv2.putText(img, "!!MOVE AWAY!!", (120, 100), cv2.FONT_HERSHEY_SIMPLEX,2, [0,0,255] , 4)
            
            face = cv2.putText(img, str(D/10) + " cm", (300, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0) , 2, cv2.LINE_AA)

        
        #Mask Detection Estimator
        for x,y,w,h in detect_face:
            rec = cv2.rectangle(img, (x,y), (w+x,h+y), (255,0,255), 4)
            fetch_face = img[y:y+h, x:x+w, :] #collecting dimensions of face only using array indexing
            fetch_face = cv2.resize(fetch_face, (50,50))
            fetch_face = fetch_face.reshape(1,-1)
            detect_mask = svc.predict(fetch_face)
            n = names[int(detect_mask)]
            cv2.putText(img, n, (x,y),font, 1, (234,0,240), 2)
            print(n)
        cv2.imshow('show',img)
        if cv2.waitKey(2) == 27: #press ESC to terminate the program
            break
video.release()
cv2.destroyAllWindows()
```
