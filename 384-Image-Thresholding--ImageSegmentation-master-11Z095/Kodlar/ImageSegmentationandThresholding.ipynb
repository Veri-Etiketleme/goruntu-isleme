# Global Thresholding

```
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('sudoku.png',0)
img = cv2.medianBlur(img,5)

ret,gloT = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
mean = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
gaus = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)

names = ['Original', 'Global','Mean', 'Gaussian']
im = [img, gloT, mean, gaus]

for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(im[i],'gray')
    plt.title(names[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

# **Otsu's Thresholding**

```
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

img = cv2.imread('shape.tif',0)

print('Histogram of normal image')
histr2 = cv2.calcHist([img],[0],None,[256],[0,256]) 
plt.plot(histr2) 
plt.show()  

# Otsu's thresholding
ret2,th2 = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
print('Image after applying otsu thresholding')
cv2_imshow(th2)

# Otsu's thresholding after Gaussian filtering
blur = cv2.GaussianBlur(img,(5,5),0)
ret3,th3 = cv2.threshold(blur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

print('Histogram of image after removing noise')
histr1 = cv2.calcHist([blur],[0],None,[256],[0,256]) 
plt.plot(histr1) 
plt.show()


print('Image after removing noise and otsu')
cv2_imshow(th3)

 
```

# **Image Segmentation using KMeans**

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Load image 
img = cv2.imread("view.jpeg")

#Converting to RGB
imgRGB=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#------Parameters for cv2.kmeans

K = 3
#Converting to vectors 
imgVector = imgRGB.reshape((-1,3))
imgVector = np.float32(imgVector)

#When to stop the iteration for the algorithm 
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)

#---------End of parameters

#Applying cv2.kmeans
comp,label,centers=cv2.kmeans(imgVector,K,None,criteria,10,cv2.KMEANS_PP_CENTERS)

# converting centers in to uint8
centers = np.uint8(centers)
#accesing labels to regenerate the clustered image
img1 = centers[label.flatten()]
# imgF is the result of the frame which has gone k means clustering
imgF = img1.reshape((imgRGB.shape))

figure_size = 15
plt.figure(figsize=(figure_size,figure_size))
plt.subplot(1,2,1),plt.imshow(imgRGB)
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(imgF)
plt.title('K = %i' % K), plt.xticks([]), plt.yticks([])
plt.show()
```
